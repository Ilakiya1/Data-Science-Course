### 1) Polymorphism

- Polymorphism will behave differently in multiple states which completely depends on the data that we are passing.

```python
def test(a,b):
    return a + b
```

```python
test(4,7)
```

```python
test("Smart", "Tv")
```

```python
test([1,2,3], [3,4,5])
```

```python
class data_science():
    def syllabus(self):
        print("This is the method for data science syllabus")
```

```python
class java():
    def syllabus(self):
        print("This is the method for java syllabus")
```

def class_parcer(class_obj):
    for i in class_obj:
        i.syllabus()

```python
obj_data_science = data_science()
obj_java = java()
class_obj = [obj_data_science, obj_java]
```

```python
class_parcer(class_obj)
```

### 2) Encapsulation:

- Encapsulation prevents direct access of a data in oops concept.
- Only the user who creates the object can access the data and others don't have access to it.
- It is used to hide the entire implementation from others who are trying to access the data.
- Hiding all the internal composition and only the high level function will be visible to the user


```python
class test1:
    def __init__(self, a,b):
        self.a = a
        self.b = b
```

```python
t = test1(4, 5)
```

```python
t.a
```

```python
t.b
```

```python
t.a = 100
t.a
```

```python
t.b = 200
t.b
```

```python
class car:
    def __init__(self, year, make, model, speed):
        self.__year = year
        self.__make = make
        self.__model = model
        self.__speed = 0
        
    def set_speed(self, speed):
        self.__speed = 0 if speed < 0 else speed
        
    def get_speed(self):
        return self.__speed
```

```python
obj_car = car(2021, "sonnet", "Kia", 12)
```

```python
obj_car._car__year
```

```python
obj_car.set_speed(-19)
```

```python
# No one can access other than the admin or the creator
obj_car.set_speed(19)
```

```python
# Only the creator can access the data with the below syntax
obj_car._car__speed
```

```python
obj_car.get_speed()
```

```python
obj_car
```

```python
class bank_account:
    def __init__(self, balance):
        self.__balance = balance
        
    def deposit(self, amount):
        self.__balance = self.__balance + amount
        
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance = self.__balance - amount
            return True
        else:
            return False
    def get_balance(self):
        return self.__balance
```

```python
obj_bank_account = bank_account(1000)
```

```python
# By using this method, we can access or see the balance amount which are private
obj_bank_account._bank_account__balance 
```

```python
# Normal user can access the account balance using get method
obj_bank_account.get_balance() 
```

```python
obj_bank_account.deposit(5000) 
```

```python
obj_bank_account.get_balance() 
```

```python
obj_bank_account.withdraw(2000)
```

```python
obj_bank_account.get_balance() 
```

### 3)Inheritance:

- A class which inherits a property from another class is called as Inheritance

```python
class parent:
    def test_parent(self):
        print("This is the parent class")

class child(parent):
    pass
```

```python
child_obj = child()
```

```python
child_obj.test_parent()
```

### 3.1 Multi-label Inheritance

```python
class class1:
    def test_class1(self):
        print("This is the class2")
        
class class2(class1):
    def test_class2(self):
        print("This is the class2")
        
class class3(class2):
    def test_class3(self):
        print("This is the class3")
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

```python
obj_class3.test_class3()
```

### 3.2 Multiple Inheritance

```python
class class1():
    def test_class1(self):
        print("This is the class1")
        
class class2():
    def test_class2(self):
        print("This is the class2")
        
class class3(class1, class2):
    pass
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```
