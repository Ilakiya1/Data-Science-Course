```python
l = [1,2,3,4,5, [7,8,9], "smith", "nila"]
```

```python
l1_int = []
l2_str = []

for i in l:
    logging.info("This is my first for loop log {}".format(l))
    logging.info("This is the value of i am logging {}".format(i))
    if type(i) == list:
        for j in i:
            logging.info("logging my j {j} and i is {i}".format(i = i, j = j))
            if type(j) == int:
                l1_int.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else:
        if type(i) == str:
            l2_str.append(i)
logging.info("This is my final result with all int {l1} and str {l2}".format(l1 = l1_int, l2 = l2_str))
```
