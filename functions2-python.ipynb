```python
# Reduce Functions

from functools import reduce
```

```python
l = [1,2,3,4,5]
```

```python
reduce(lambda x, y : x+y , l)
```

```python
# Another way of using reducing function with normal function

def add(x, y):
    return x+y

reduce(add, l)
```

```python
reduce(lambda x,y : x if x>y else y, l)
```

```python
# list of even and odd numbers using Filter Function

l1 = [1,2,3,4,5,6,7]
l1

list(filter(lambda x : x % 2 == 0, l))
```

```python
list(filter(lambda x : x%2 !=0, l))
```

```python
# Filter positive and negative numbers

l3 = [1,-3,4,-1,9,-3]
l3
```

```python
list(filter(lambda x : x>=0, l3))
```

```python
list(filter(lambda x: x<=0, l3))
```

```python
l5 = ['moon', 'star', 'sun', 'eclipse']

list(filter(lambda x: len(x) < 6, l5))
```
