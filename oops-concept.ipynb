```python
# Constructor or inbuilt method

class course1:
    
    def __init__(self, phone_num, email_id, student_id):
        self.phone_num = phone_num
        self.email_id = email_id
        self.student_id = student_id
    
    def return_student_details(self):
        return self.phone_num, self.email_id, self.student_id
```

```python
name = course1(782342423, "john@gmail.com", 101)
```

```python
name.return_student_details()
```

```python
name.email_id
```

```python
name.phone_num
```

```python
name.student_id
```

```python
# Self is not a keyword, its just the reference which we can pass anything in the place of self
# Example

class course2:
    
    def __init__(new, phone_num, email_id, student_id):
        new.phone_num = phone_num
        new.email_id = email_id
        new.student_id = student_id
    
    def return_student_details(new):
        return new.phone_num, new.email_id, new.student_id
```

```python
name = course2(98766453, 'rebaca@gmail.com', 102)
name.student_id
```

```python
name.return_student_details()
```
