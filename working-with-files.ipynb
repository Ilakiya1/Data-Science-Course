### Buffered read and write

```python
import io
```

```python
with open("text.txt", "wb") as f:
    file = io.BufferedWriter(f)
    file.write(b"This is my buffered write")
    file.write(b"This is my second line")
    file.flush()
```

```python
with open("text.txt" , "rb") as f:
    file = io.BufferedReader(f)
    data = file.read()
    print(data)
```
