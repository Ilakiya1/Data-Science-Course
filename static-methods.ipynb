```python
class course1:
    def student_details(self, name, email_id, number):
        print(name, email_id, number)
        
    @staticmethod
    def mentor_class(list_mentor):
        print(list_mentor)
    
    def mentor(self, mentor_list):
        print(mentor_list)
```

```python
obj_course1 = course1()
obj_course1.mentor(["jony", "bob"])
```

```python
obj_course1.mentor_class(["kia", "sudha"])
```

```python
# By using static method, we can directly call the mentor_class function
# through the class name

course1.mentor_class(["sudha", "kia"])
```

```python
class course2:
    def student_details(self, name, email_id, number):
        print(name, email_id, number)
   
    @staticmethod     # static method is available accross all the classess
                      # static method is used to reduce memory utilization
    def mentor_mail_id(m_mail_id):
        print(m_mail_id)
    
    @staticmethod
    def mentor_class(list_mentor):
        print(list_mentor)
        course2.mentor_mail_id(["sudha@gmail.com", "kia@gmail.com"])
    
    @classmethod
    def class_name(cls, class_name):
        cls.mentor_class(["john", "bobby"])
    
    def mentor(self, mentor_list):
        print(mentor_list)
        self.mentor_class(["sudha", "kia"])
```

```python
obj_course2 = course2()
obj_course2.student_details("josh", "josh@gmail.com", 9724634571)
```

```python
obj_course2.mentor_mail_id(["sudha@gmail.com", "kia@gmail.com"])
```

```python
obj_course2.class_name("Data science course")
```

```python
obj_course2.mentor_mail_id(["sudha@gmail.com", "kia@gmail.com"])
```

### Property Decorators

```python
class course3:
    def __init__(self, course_price, course_name):
        self.__course_price = course_price    # Creating a private variable
        self.course_name = course_name
```

```python
obj = course3(5000, "Data Science")
```

```python
obj.course_name
```

```python
obj._course3__course_price   # Accessing the private variable
```

### How to make these private variable(class property) accesseble to the outer world

```python
##  __(Double Underscore) Private variable
## _(Single Underscore) proteccted variable 

class course4:
    def __init__(self, course_price, course_name):
        self.__course_price = course_price    # Creating a private variable
        self.course_name = course_name
    
    # With the help of property, private variable can be accessed by the outer world other than the creator
    @property
    def course_price_access(self):
        return self.__course_price
    
    # With the help of setter, user can set or modify a variable
    @course_price_access.setter
    def course_price_set(self, price):
        if price <= 5000:
            pass
        else:
            self.__course_price = price
     
    # With the help of delete, user can delete a variable
    @course_price_access.deleter
    def course_price_delete(self):
        del self.__course_price
```

```python
obj3 = course4(5000, "Data Science")
```

```python
obj3.course_name
```

```python
obj3.course_price_access
```

```python
obj3.course_price_set = 6000
```

```python
obj3.course_price_access
```

```python
del obj3.course_price_delete
```
