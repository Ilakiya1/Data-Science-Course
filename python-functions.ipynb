```python
def test5(a):
    n = []
    for i in a:
        if type(i) == int or type(i) == float:
            n.append(i)
    return i

test5(l)
```

```python
def test6(a):
    n = []
    for i in a:
        if type(i) == list:
            for j in i:
                if type(j) == int or float:
                    n.append(j)
        elif type(i) == int or type(i) == float:
            n.append(i)
    return n

test6(l)
```

```python
# * args represents that any number of arguments can be passed
# Everything (i,e) the result will be inside a tuple

def fun(*args):
    return args

fun(3)
```

```python
fun(1,2,3,4.5,[7,8,9], (8,9))
```

```python
def fun2(*args, a):
    return args, a

fun2(1,2345,64, a = "name")
```

```python
# Fibonacci series:

def fib(n):
    a, b = 0,1
    for i in range(n):
        yield a
        a,b = b, a+b

fib(10)
        
```

```python
for i in fib(10):
    print(i)
```
