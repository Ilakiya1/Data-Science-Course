### Custom Exception Handling

```python
Age = int(input("Enter your age:"))
```

```python
# The above age is not valid, so we are using our own exception handling

class validateage(Exception):   # Inheriting the exception class
    def __init__(self, msg):
        self.msg = msg
```

```python
def validate_age(age):
    if age < 0:
        raise validateage("Age should not be less than zero")
    elif age > 200:
        raise validateage("Age is too high, Not a valid age")
    else:
        print("Age is valid")
```

```python
try:
    age = int(input("Enter your age:"))
    validate_age(age)
except validateage as e:
    print(e) 
```

### List of General Use Exceptions

```python
try:
    a = 10
    10 / 0
except ZeroDivisionError as e:
    print(e)
```

```python
# Value error or type error

try:
    int("kia")
except (ValueError, TypeError) as e:
    print(e)
```

```python
try:
    int("kia")
except (ValueError, TypeError) as e:
    print("This will catch an error", e)
```

```python
try:
    import new
except ImportError as e:
    print(e)
```

```python
try:
    d = {1: [3, 4, 5, 6], "Key": "value"}
    d["key10"]
except KeyError as e:
    print(e)
```

```python
try:
    "kia".test()
except AttributeError as e:
    print(e)
```

```python
try:
    l = [1,2,4,5]
    l[20]
except IndexError as e:
    print(e)
```

```python
try:
    123 + "kia"
except TypeError as e:
    print(e)
```

```python
try:
    with open("test2.txt", 'r') as f:
        f.read()
except FileNotFoundError as e:
    print(e)
```
