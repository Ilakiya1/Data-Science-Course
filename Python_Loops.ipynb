## Looping Statements
1. While loop
2. For loop
3. Nested loops
4. Loop Controls (break, continue, pass)

```python
# while loop - with while - else block

#retirement_age = 60
joining_age = 25

while joining_age <= 60:
    joining_age = joining_age +1
    print(joining_age)
else:   
    print("Its time to retire")
```

```python
amount = 1000

while amount >= 950:
    print(amount)
    amount = amount-1
    
print("Finished")
```

```python
total_amount = 1000

while total_amount != 0: 
    #print(total_amount)
    total_amount = total_amount - 100
    print(total_amount)
else:
    print("Insuffiecient balance")
```

## For loop



```python
Container = ['name', 1,'e',3,4,'e', 'banana', 'apple']
Container
```

```python
type(Container)
```

```python
Container[3]
```

```python
for x in Container:
    print(x)
```

```python
for x in Container:
    print(x, end = "\n")
    if x == 'banana':
        print("Present inside the basket")
        break
    else:
        print("Not present inside")
```

```python
fruit = 'mango'
for x in fruit:
    print(x)
```

```python
name = 'ilakiya'
for i in name:
    print(i, end ='')
```

```python
# Range - For creating patterns
for i in range(1,6,2):
    print(i)
```

```python
for i in range(1,10):
    print(i)
```

```python
## Nested loops
n = 5

for i in range(0, n):
    for j in range(0, i+1):
        print("*", end = '')
    print('\r')
```

```python
for i in Container:
    if i == 'e':
        print('Present')
        break
    print(i)
```

```python
lis = [1,2,3],[3,4,6],[6,7,8]
lis

for i in lis:
    for j in i:
        print(j, end ="")
```
