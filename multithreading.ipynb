### Multithreading

- When a program executes multiple times in a single processor or core, then multithreading comes into picture where each program gets executed partially or consecutively.

- Multiple programs/functions that are executed consecutively within a single core or processor is called as multithreading.

```python
import threading

def test(id):
    print("This is my test id %d" % id)
```

```python
test(10)
```

```python
test(5)
```

```python
test(2)
```

### With the help of a thread, we can execute all the above 3 different functions together simultaneaously

```python
thread = [threading.Thread(target = test, args = (i,)) for i in [10, 5, 2]]
```

```python
thread
```

```python
for t in thread:
    t.start()
```

```python
import urllib.request

def file_download(url, filename):
    urllib.request.urlretrieve(url, filename)
```

```python
file_download("https://github.com/Ilakiya1/Data-Science-Course/blob/main/file1.txt","test1.txt")
```

```python
url_list = ['https://github.com/Ilakiya1/Data-Science-Course/blob/main/file1.txt', 'https://github.com/Ilakiya1/Data-Science-Course/blob/main/file2.txt']

url_list
```

```python
data_file_list = ['data1.txt', 'data2.txt']
data_file_list
```

```python
thread1 = [threading.Thread(target = file_download, args = (url_list[i], data_file_list[i])) for i in range(len(url_list))]
```

```python
for t in thread1:
    t.start()
```

```python
import time 

def test2(x):
    for i in range(10):
        print("print the value of x %d and print the value of i %d " %(x,i))
        time.sleep(1)
```

```python
thread2 = [threading.Thread(target = test2, args = (i,)) for i in [100, 10, 20, 5]]
```

```python
for t in thread2:
    t.start()
```

```python
test2(10)
```

```python
test2(10)
```

```python
shared_var = 0
lock_var = threading.Lock()
def test3(x):
    global shared_var
    with lock_var:
        shared_var = shared_var + 1
        print("value of x %d and value of shared_var %d" %(x, shared_var))
        time.sleep(1)
        
thread4 = [threading.Thread(target = test3, args = (i,)) for i in [1,2,3,4]]
```

```python
for t in thread4:
    t.start()
```

```python
test3(1)
```

```python
test3(2)
```
